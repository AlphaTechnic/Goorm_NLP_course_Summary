10. $v_1,v_2$가 matrix $A$의 서로 다른 eigenvalues $\lambda_1,\lambda_2$에 해당하는 eigenvectors라면, $v_1$과 $v_2$는 linearly independent하다.

A. True


distinct eigenvalues $\lambda_1, \lambda_2$를 가정하고, 각 eigenvalue에 대응하는 eigenvector들을 v1, v2라고 하자.

그리고, c1v1 + c2v2 = 0 where c1, c2 implies c1 = c2 = 0을 보인다.

위 식에 A를 곱하면, 

c1$\lambda_1$v1 + c2$\lambda_2$v2 = 0

똑같은 식에 $\lambda_1$을 곱하면,

c1$\lambda_1$v1 + c2$\lambda_1$v2 = 0

유도된 두 식을 빼면

c2($\lambda_2$ - $\lambda_1$)v2 = 0

v2는 eigenvector여서 0벡터일 수 없고, distinct eigenvalues $\lambda_1, \lambda_2$를 가정했으므로, c2 = 0일수 밖에 없다. 마찬가지 방법으로 c1 = 0을 유도한다.

11. $A$가 symmetric matrix라면, $A$의 모든 eigenvector는 서로 orthogonal 하다.

A. ~~True~~ False

Av1 = $\lambda_1$v1...........1

Av2 = $\lambda_2$v2...........2

1 식에 $v_2^T$를 곱하고 2 식에 $v_1^T$를 곱한다.

$v_2^T$Av1 = $v_2^T$$\lambda_1$v1...........1'

$v_1^T$Av2 = $v_1^T$$\lambda_2$v2...........2'

1'의 양변을 transpose한 뒤 2'와 빼면,

0 = ($\lambda_1 - \lambda _2$)$v_1^T$v2

$v_1^T$v2=0 iff v1 is orthogonal to v2



그런데 , ($\lambda_1 - \lambda _2$)가 0 이 아니라면 , False

(9번의 반례를 생각해보면 됨. [1, 2], [2, 1]도 훌륭한 eigenvector이지만, 서로 orthogonal하지는 않다.)
